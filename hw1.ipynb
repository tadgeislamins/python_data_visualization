```python
x_stats1 = df[(df.Son >= 66) & (df.Son < 68)].Father
x_stats2 = df[(df.Son >= 68) & (df.Son < 70)].Father
x_stats3 = df[(df.Son >= 70) & (df.Son < 72)].Father

y_stats1 = df[(df.Father >= 64) & (df.Father < 67)].Son
y_stats2 = df[(df.Father >= 67) & (df.Father < 69)].Son
y_stats3 = df[(df.Father >= 69) & (df.Father < 72)].Son
```

```python
plt.rcParams['font.family'] = 'cursive'

prop = font_manager.FontProperties(fname='13189-fontps.ttf')
```

```python
plt.figure(figsize=(12,9))
grid = plt.GridSpec(10, 10)
general_ax = plt.subplot(grid[0:4, 0:4])
y_stats1_ax = plt.subplot(grid[0:4, 4:6])
x_stats1_ax = plt.subplot(grid[4:6, 0:4])
y_stats2_ax = plt.subplot(grid[0:4, 6:8])
x_stats2_ax = plt.subplot(grid[6:8, 0:4])
y_stats3_ax = plt.subplot(grid[0:4, 8:])
x_stats3_ax = plt.subplot(grid[8:, 0:4])

general_ax.scatter(x, y)
general_ax.set_xticks(range(58, 81, 2))
general_ax.set_yticks(range(58, 81, 2))
general_ax.set_xlabel("Father's height (inches)", fontdict={'fontsize': 10})
general_ax.set_ylabel("Son's height (inches)", fontdict={'fontsize': 10})
general_ax.xaxis.tick_top()
general_ax.xaxis.set_label_position('top')

y_stats1_ax.hist(y_stats1, bins=21, orientation='horizontal')
y_stats1_ax.set_title('64 to 67 inches fathers', fontdict={'fontsize': 11})
x_stats1_ax.hist(x_stats1, bins=23)
x_stats1_ax.invert_yaxis()
x_stats1_ax.set_title('66 to 68 inches sons', fontdict={'fontsize': 11})

y_stats2_ax.hist(y_stats2, bins=21, orientation='horizontal')
y_stats2_ax.set_title('67 to 69 inches fathers', fontdict={'fontsize': 11})
x_stats2_ax.hist(x_stats2, bins=23)
x_stats2_ax.invert_yaxis()
x_stats2_ax.set_title('68 to 70 inches sons', fontdict={'fontsize': 11})

y_stats3_ax.hist(y_stats3, bins=21, orientation='horizontal')
y_stats3_ax.set_title('69 to 72 inches fathers', fontdict={'fontsize': 11})
x_stats3_ax.hist(x_stats3, bins=23)
x_stats3_ax.invert_yaxis()
x_stats3_ax.set_title('70 to 72 inches sons', fontdict={'fontsize': 11})

plt.suptitle("Distribution and correlation between fathers' and sons' height\n\
(based on an experiment by Karl Pearson)", fontproperties=prop, fontsize=25)
plt.tight_layout()
plt.show()
```
